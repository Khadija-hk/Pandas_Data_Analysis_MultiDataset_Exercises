DATA ANALYSIS

```python
df = pd.read_csv("video_games.csv")
df = df.dropna()
df
```

```python
print(df.head(5))
print(df.columns)
```

```python
print(df.describe())
```

```python
#Group the sales data of each region on Year_of_Release and plot the same
sales_by_year = df.groupby('Year_of_Release')[['NA_Sales', 'EU_Sales', \
                        'JP_Sales', 'Other_Sales', 'Global_Sales']].sum()

# Plot the sales data for each region over the years
plt.figure(figsize=(12, 6))
sales_by_year.plot(kind='line', marker='o', linestyle='-')
plt.title('Sales Data by Region Over the Years')
plt.xlabel('Year of Release')
plt.ylabel('Sales (in millions)')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
#Group North American, Japan, European & Other region sales data as per Genre and plot it
sales_by_genre = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Plot the results
plt.figure(figsize=(12, 6))
sales_by_genre.plot(kind='bar', stacked=True)
plt.title('Sales by Genre and Region')
plt.xlabel('Genre')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

```python
#Group North American, Japan, European & Other region sales data as per Platform and plot it
sales_by_platform = df.groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Plot the results
plt.figure(figsize=(12, 6))
sales_by_platform.plot(kind='bar', stacked=True)
plt.title('Sales by Platform and Region')
plt.xlabel('Platform')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

```python
#Plot the Sales in all regions as per Platform
sales_data = df[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

# Plot boxplot for each region as per platform
plt.figure(figsize=(12, 6))
sales_data.boxplot(by='Platform', rot=45)
plt.title('Sales Distribution by Platform')
plt.xlabel('Platform')
plt.ylabel('Sales')
plt.show()
```

```python
#Find out which Genre games are sold in highest numbers
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Genre', order=df['Genre'].value_counts().index)
plt.title('Number of Games Sold by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Games Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#global sales as per Genre
sales_genre = df.groupby('Genre')['Global_Sales'].sum().reset_index()

# Plot a barplot to visualize global sales by genre
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_genre, x='Genre', y='Global_Sales')
plt.title('Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```
